# 互信息法 mutual_info

```python
import pandas as pd
from sklearn.feature_selection import mutual_info_classif as MI
```

```python
# 一、利用阈值筛选
data = pd.read_csv('../breast_cancer_m.csv')
x = data.iloc[:, :-1]
y = data['label']
x.shape
```

```python
MI_result=MI(x,y) #实例化互信息法得出每个特征对结果的贡献度
MI_result
```

```python
MI_result>0.1#提取大于0.1的特征的bool数组
```

```python
x_MI = x[x.columns[MI_result > 0.1]]  # 提取大于0.1的特征
x_MI
```

```python
# 二、利用SelectKBest类 筛选互信息法前k个最优特征

from sklearn.feature_selection import SelectKBest

SKB=SelectKBest(score_func=MI,k=15) #实例化 筛选互信法前15个最优特征
SKB.fit(x,y)
SKB.scores_#得到所有特征贡献度评分  fit后就可以调用scores_,不能在transform后调用
```

```python
SKB.get_support()#得到所有特征的bool数组  true为被筛选出来的15个特征中的一个
```

```python
x_MI_K15=SKB.transform(x)#得到筛选的15个特征列
x_MI_K15
```

```python
pd.DataFrame(x_MI_K15,columns=x.columns[SKB.get_support()])
```
