```python
#! pip3 install pyradiomics

from radiomics import featureextractor
import os
```

```python
#一、影像特征提取（不进行预处理参数设置的简单例子）

basePath='/Users/chenyu/Desktop/Medical paper/test/brain1'
imageFile=os.path.join(basePath,'brain1_image.nrrd')
maskFile=os.path.join(basePath,'brain1_label.nrrd')
extractor=featureextractor.RadiomicsFeatureExtractor() #实例化提取器
featureVector=extractor.execute(imageFile,maskFile) #执行提取器得到特征向量
for featureName in featureVector.keys(): #对特征向量中每个特征遍历输出
    print('%s: %s'%(featureName,featureVector[featureName]))
```

```python
# 二、常用参数设置（使用SimpleITK包）  图像预处理
import SimpleITK as sitk

basePath2 = '/Users/chenyu/Desktop/Medical paper/test/brain1'
imageFile2 = os.path.join(basePath2, 'brain1_image.nrrd')
maskFile2 = os.path.join(basePath2, 'brain1_image.nrrd')
settings={} #参数设置
settings['binWidth']=25 #宽度设为25
settings['resampledPixelSpacing']=[3,3,3] # 重采样像素间隔，推荐三个方向一样数字
settings['interpolator'] = sitk.sitkNearestNeighbor  # 差值方法 这里采用的是SimpleITK的最近邻差值方法 代号为1 见下方详解
settings['normalize']=True #使用标准化
extractor2=featureextractor.RadiomicsFeatureExtractor(**settings)#**代表传入的这个参数中的数按个数传入
featureVector2=extractor2.execute(imageFile2,maskFile2)
for featureName2 in featureVector2.keys():
    print('%s : %s'%(featureName2,featureVector2[featureName2]))

```

```python
# 三、个别特征提取（不提取影像的全部特征）
extractor.disableAllFeatures() #1.先关闭所有特征
extractor.enableFeaturesByName(firstorder=['Mean','Skewness']) #2.再开启所需要的特征

featureVector3=extractor.execute(imageFile,maskFile)
for featureName3 in featureVector3.keys():
    print('%s : %s'%(featureName3,featureVector3[featureName3]))
```

```python
# 4.选取某种类型的特征
extractor.disableAllFeatures()
extractor.enableFeatureClassByName('glrlm')
extractor.enableFeatureClassByName('glcm')

featureVector4=extractor.execute(imageFile,maskFile)
for featureName4 in featureVector4.keys():
    print("%s : %s"%(featureName4,featureVector4[featureName4])) 
```

```python
# 五、滤波器选择
extractor.enableImageTypes(Original={},LoG={'sigma':[4.0]},Wavelet={}) #滤波器参数设置  

featureVector5=extractor.execute(imageFile,maskFile)
for featureName5 in featureVector5.keys():
    print("%s : %s"%(featureName5,featureVector5[featureName5]))
```

```python
# 6、批量提取影像组学特征
import pandas as pd

basePath6 = '/Users/chenyu/Desktop/Medical paper/test/featureExtraction'
folders = os.listdir(basePath6)#列出文件夹中的所有文件
folders
```

```python
df=pd.DataFrame()
extractor6=featureextractor.RadiomicsFeatureExtractor()
for folder in folders:
    files = os.listdir(os.path.join(basePath6, folder))  # 列出该文件夹中的所有文件
    print(files)
    for file in files:
        if file.endswith('image.nrrd'): #如果该文件已image.nrrd结尾 则是影像文件
            imageFile6=os.path.join(basePath6,folder,file)
        if file.endswith('label.nrrd'):
            # 如果该文件已label.nrrd结尾 则是roi文件
            maskFile6 = os.path.join(basePath6, folder, file)
    featureVector6=extractor6.execute(imageFile6,maskFile6)
    df_new=pd.DataFrame.from_dict(featureVector6.values()).T #从这个字典转化为数据帧，但因为每行是特征 所以要转置
    df_new.columns=featureVector6.keys()#给数据帧取列名（即每列的特征名字）
    df=pd.concat([df,df_new])#拼接到总的数据帧上
df.to_excel(os.path.join(basePath6,'py批量提取影像特征结果.xlsx'))#存为Excel文件
```
