```python
# 一、归一化处理（最大最小值处理，Min-Max Scaling） 特征值范围划到0-1之间（可以改） 服从正态分布
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd

dataFile='/Users/chenyu/Desktop/Medical paper/test/breast_cancer_m.csv'
data=pd.read_csv(dataFile)
data.head()
```

```python
x=data.iloc[:,:-1]
y=data['label']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
scaler=MinMaxScaler()#可以加入参数feature_range=(,)修改归一化后的范围
scaler=scaler.fit(x_train)#也可以使用partial_fit(x_trian)，partial_fit适合数据集密的情况
x_train_scaler=scaler.transform(x_train)#对于训练集，可以直接使用fit_transform(x_train)得到归一化结果
x_train_scaler_df=pd.DataFrame(x_train_scaler,columns=x_train.columns)
x_train_scaler_df 
```

```python
x_test_scaler=scaler.transform(x_test)#测试集不能直接使用fit_transform(x_test),必须经过测试集实例化后的fit(x_train)，使用transform(x_test)
x_test_scaler_df=pd.DataFrame(x_test_scaler,columns=x_test.columns)
x_test_scaler_df
```

```python
# 二、标准化处理（standardization ，Z-score normalization处理），服从标准正态分布，范围在（-1，-1）
from sklearn.preprocessing import StandardScaler
z_scaler=StandardScaler()
z_scaler_train=z_scaler.fit(x_train)
x_train_z_scaler=z_scaler_train.transform(x_train)
x_train_z_scaler_df=pd.DataFrame(x_train_z_scaler,columns=x_train.columns)

```

```python
x_test_z_scaler = z_scaler_train.transform(x_test)
x_test_z_scaler_df = pd.DataFrame(x_test_z_scaler, columns=x_train.columns)
x_test_z_scaler_df
```
